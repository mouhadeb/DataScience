```r
rmse <- function(y,y_hat) {
    err <- sqrt(sum((y_hat-y)^2,na.rm=T)/length(y))
    return(err)
}

rmsle <- function(y,y_hat) {
    err <- sqrt(sum((log(y_hat+1)-log(y+1))^2,na.rm=T)/length(y))
    return(err)
}
```

```r
err_res <- NULL
```

```r
mod1 <- lm(temperature ~., data=train)
summary(mod1)
```

```r
pred1 <- predict(mod1,newdata=test)
pred2 <- predict(mod1,newdata=train)
rmse(test$temperature,pred1)
rmsle(test$temperature,pred1)
rmse(train$temperature,pred2)
rmsle(train$temperature,pred2)
err_res <- rbind(err_res, data.frame(Name="Base Linear regression", Model="mod1", 
                                     test_RMSE=rmse(test$temperature,pred1), 
                                     test_RMSLE=rmsle(test$temperature,pred1),
                                   train_RMSE=rmse(train$temperature,pred2), 
                                     train_RMSLE=rmsle(train$temperature,pred2)))
err_res
```
