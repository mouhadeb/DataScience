```r
### The error we will use is the RMSE and RMSLE. 
rmse <- function(y,y_hat) {
    err <- sqrt(sum((y_hat-y)^2,na.rm=T)/length(y))
    return(err)
}

rmsle <- function(y,y_hat) {
    err <- sqrt(sum((log(y_hat+1)-log(y+1))^2,na.rm=T)/length(y))
    return(err)
}
```

```r
### Table of resulting errors
### Name, Model, RMSE, RMSLE
err_res <- NULL
```

## Linear Models

```r
## model with only the original variables. 
mod1 <- lm(cnt ~., data=train)
summary(mod1)
```

```r
# "Adjusted R-squared:  0.7485 " this model explane only 0.04 of the vaiables
```

```r
pred1 <- predict(mod1,newdata=test)
rmse(test$cnt,pred1)
rmsle(test$cnt,pred1)
err_res <- rbind(err_res, data.frame(Name="Base Linear regression", Model="mod1", # here we crate a new dataframe that contain the result of the model
                                     RMSE=rmse(test$cnt,pred1), 
                                     RMSLE=rmsle(test$cnt,pred1)))
```

```r
pred1 <- predict(mod1,newdata=train)
rmse(train$cnt,pred1)
rmsle(train$cnt,pred1)
err_res <- rbind(err_res, data.frame(Name="Base Linear regression", Model="mod1", # here we crate a new dataframe that contain the result of the model
                                     RMSE=rmse(train$cnt,pred1), 
                                     RMSLE=rmsle(train$cnt,pred1)))
```

```r
#Treating the numeric vars by categorize using the IQR- q25 and q75
IQR <- function(x) {
  q25 <- quantile(x,probs = 0.25, na.rm=T)
  q75 <- quantile(x,probs = 0.75, na.rm=T)
  iqr <- paste(q75-q25, " (",q25," , ",q75,")",sep="")
    print(paste(q25,'q25 is ', sep=' '))
    print(paste(q75,'q75 is ', sep=' '))
    print(paste(q75-q25,'IQR is ', sep=' '))
  return(iqr)
}
```
