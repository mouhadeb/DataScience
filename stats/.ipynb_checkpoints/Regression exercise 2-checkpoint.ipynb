## Linear Regression

```r
tips <- read.csv("../data/tips.csv")
head(tips)
```

```r
summary(tips)
```

```r
hist(tips$tip)
```

```r
mod2 <- lm(tip ~ total_bill + size + sex + smoker + day + time , data=tips)
summary(mod2)
```

```r
vif(mod2)
```

```r
Table2(mod2) # hear we look at the relative risk and not at the Betott
```

```r
Table2.forestplot(mod2)
```

```r
pred <- predict(mod2,newdata=tips)
hist(pred)
```

```r
scatter.smooth(pred,tips$tip) #x is how mach tips left in the real and the y is the prediction
```
