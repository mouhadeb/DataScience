```r
zm1 <- movies_ff_v %>% select("revenue"
         ,"popularity"
         ,"runtime"
         ,"budget"
         ,"seasonality"
         ,"actor0_prev_revenue"
         ,"actor1_prev_revenue"
         ,"actor2_prev_revenue")
```

```r
zm2 <- movies_ff_v %>% select("revenue","sw_lang_en"
         ,"sw_web_presence"
         ,"sw_has_poster"
         ,"sw_tagline"
                                  ,"high_release_month"
                                    ,"lang_US"
                                    ,"lang_FR"
                                    ,"lang_RU"
                                    ,"lang_ES"
                                    ,"lang_JA")
```

```r
library(Hmisc)

res2 <-  rcorr(as.matrix(zm1),type="spearman")
res2
```

```r
library(corrplot)

corrplot(res2$r, type="upper", order="hclust", 
         p.mat = res2$P, sig.level = 0.01, insig = "blank")
```
